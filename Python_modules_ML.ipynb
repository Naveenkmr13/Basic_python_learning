#NUMPY LEARNING

1) CHECKING TIME TAKEN FOR LIST AND NUMPY

TIME FOR LIST #**0.00011698500000001388**

```python
import numpy as np #0.0001343380000005112
from time import process_time

arr=[i for i in range(1,101)]

st=process_time()

brr=[i+2 for i in arr]

end=process_time()
print(brr)
print(st)
print(end)
print(end - st) #0.00011698500000001388
```

TIME FOR NUMPY IS  **0.0001528569999997842**

```python
np_array=np.array([i for i in range(1,101)])
print(np_array)
st=process_time()
np_arr=np.array([i+2 for i in np_array])
end=process_time()
print(np_arr)
print(end-st) #0.0001528569999997842
```

**1)LINE PLOT**

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)

# plt.figure(figsize=(10,6))
plt.title('SINE WAVE')
plt.plot(x,y)
plt.show()

#IF YOU WANY CURVE PLOT THEN YOU NEED TO USE SMALL RANGE LARGE VALUES
```

```python
plt.plot(x,z)
plt.title('cosine wave')
plt.show
```

```python
#parabola wave
x = np.linspace(-10,10,20)
y = x**2


plt.title("parabola wave")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y,'.',color='green')   #or plt.plot(x,y,'r+')  or 'g.'
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.grid(True,alpha=0.5) #alpha is thickness
plt.show()
```

multiple plots in single figure

```python
x = np.linspace(0,10,100)
plt.title('multiple')
plt.plot(x,np.sin(x),'r--')
plt.plot(x,np.cos(x),'g.')
plt.show()
```

**2)BAR PLOT**

basic plot

```python
X = [100,50,150,40,70]
Y = ['ENGLISH','FRENCH','SPANISH','LATIN','GERMAN']
plt.bar(Y,X)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.grid(True,alpha=0.5) #alpha is thickness
plt.show()
```

create own figure

```python
fig = plt.figure()
ax = fig.add_axes([0,0,0.5,0.5]) #x-axis,y-axis,width,height
lang = ['ENGLISH','FRENCH','SPANISH','LATIN','GERMAN']
people = [100,50,150,40,70]
ax.bar(lang,people)
plt.xlabel('LANGUAGES')
plt.ylabel('PEOPLE')
plt.title('BARPLOT',fontsize=17)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.grid(True,alpha=0.5) #alpha is thickness
plt.show()
```

**3)PICCHART PLOT**

```python
import matplotlib.pyplot as plt
fi = plt.figure()
axs = fi.add_axes([0,0,0.5,0.5])
lang = ['ENGLISH','FRENCH','SPANISH','LATIN','GERMAN']
people = [100,50,150,40,70]
axs.pie(people,labels=lang,autopct='%1.1f%%')

plt.show()
```

**4)SCATTERPLOT**

```python
import numpy as np
f=plt.figure()
ai=f.add_axes([0,0,0.5,0.5])
x = np.linspace(0,10,10)
y = np.sin(x)
z = np.cos(x)
ai.scatter(x,y,s=23) # 3 attribute must last color(c) or size(s)
ai.scatter(x,z,c='r')
plt.show()
```

**3D PLOTS**

1.LINE 3D

```python

fig = plt.figure()
ax = fig.add_subplot(111,projection="3d") # 111 => (number of rows,number of columns,position index of the subplot)
x = np.linspace(0,100,50)
y = np.sin(x)
z = np.cos(x)
ax.plot(x,y,z,'--',label='3D line')
plt.show()
```

2.SCATTER 3D

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection="3d") # 111 => (number of rows,number of columns,position index of the subplot)
x = 20*np.random.random(100)
print(x)
y = np.sin(x)
z = np.cos(x)
ax.scatter(x,y,z,color="r")
plt.show()
```

5)Distribution plot

finding the range of the particular column.

A distribution plot shows how a single variable is spread across its range of values  like a histogram, curve, or both.

```python
import pandas as pd
import seaborn as sns
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
df = pd.DataFrame(housing.data,columns=housing.feature_names)
df['Price'] = housing.target
print(df.head())

sns.displot(data=df,x='Price',kde='True',color='r',rug='True',alpha=0.4) #alpha is thickness of the bar (>0<1)
# sns.displot(df['Price'],kind="kde") # or (data["HouseAge"]) kind is line,scatter etc..
```

corelation concept


1.   positive
2.   negative





1.   annot tells Seaborn to write the numerical values inside the cells of a heatmap.
2.   annot_kws  is a dictionary of keyword arguments passed to the text labels (annotations).

'size': 4  sets the font size of those numbers to 4 (very small).


1.   Cbar the legend for colors, showing what the colors represent (usually numerical values)
2.   A color bar appears on the right side.

It shows the color scale  e.g., darker = higher values.





```python
import matplotlib.pyplot as plt
cor = df.corr()
plt.figure(figsize=(10,6))
sns.heatmap(data=cor,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='coolwarm')

```
